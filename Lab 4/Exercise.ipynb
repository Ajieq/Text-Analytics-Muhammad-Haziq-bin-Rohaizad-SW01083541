```python
import pandas as pd
import re
import string
from bs4 import BeautifulSoup

# Load dataset (handle encoding like Example 1)
file_path = "UNITENReview.csv"
df = pd.read_csv(file_path, encoding="latin1")

# Display column content without truncation (optional)
pd.set_option("display.max_colwidth", None)
```

```python
def convert_to_lowercase(text):
    return str(text).lower()

df["lowercased"] = df["Review"].apply(convert_to_lowercase)
```

```python
def remove_urls(text):
    return re.sub(r"http\S+|www\S+", "", str(text))

df["urls_removed"] = df["lowercased"].apply(remove_urls)
```

```python
def remove_html_tags(text):
    return BeautifulSoup(str(text), "html.parser").get_text()

df["html_removed"] = df["urls_removed"].apply(remove_html_tags)
```

```python
def remove_emojis(text):
    return re.sub(r"[^\x00-\x7F]+", "", str(text))

df["emojis_removed"] = df["html_removed"].apply(remove_emojis)
```

```python
def remove_punctuation(text):
    return str(text).translate(str.maketrans("", "", string.punctuation))

df["punctuations_removed"] = df["emojis_removed"].apply(remove_punctuation)
```

```python
def remove_numbers(text):
    return re.sub(r"\d+", "", str(text))

df["numbers_removed"] = df["punctuations_removed"].apply(remove_numbers)
```

```python
stop_words = {
    "a","an","the","and","or","but","if","while","is","am","are","was","were",
    "be","been","being","have","has","had","do","does","did","in","on","at",
    "to","from","for","of","with","as","by","about","into","over","after",
    "before","between","out","against","during","without","within","along",
    "across","this","that","these","those","it","its","i","you","he","she",
    "they","we","my","your","his","her","their","our","me","him","them"
}
```

```python
def remove_stopwords(text):
    words = str(text).split()
    filtered_words = []
    for word in words:
        lower_word = word.lower()
        if lower_word not in stop_words:
            filtered_words.append(word)
    return " ".join(filtered_words)

df["stopwords_removed"] = df["numbers_removed"].apply(remove_stopwords)
```

```python
df["Cleaned_Review"] = df["stopwords_removed"]
```
